```python
from tensorflow import keras
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('insurance_data.csv')
```

```python
df.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(df[['age','affordibility']],df['bought_insurance'],random_state=25)
```

```python
x_train_scale = x_train.copy()
x_test_scale = x_test.copy()
x_train_scale['age'] = x_train['age']/100
x_test_scale['age'] = x_test['age']/100
```

```python
x_test_scale
```

```python
model = keras.Sequential([
            keras.layers.Dense(1,input_shape=(2,),activation='sigmoid',kernel_initializer='ones',bias_initializer='zeros')
])

model.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics=['accuracy']
             )

model.fit(x_train_scale,y_train,epochs=5000)
```

```python
model.evaluate(x_test_scale,y_test)
```

```python
model.predict(x_test_scale)
```

```python
class CNN:
    def __init__(self):
        self.w1 = 1
        self.w2 = 1
        self.bias = 0
    
    def log_less(y_true,y_pred):
        return -np.mean((y_true)*np.log(y_pred) + (1-y_true)*np.log(1-y_pred))

    def np_sigmoid(self,x):
        return 1/(1+np.exp(-x))
    
    def predict(self,x):
        total_sum = self.w1*x['age'] + self.w2*x['affordibility'] + bias
        return self.np_sigmoid(total_sum)
    
    def fit(self,x,y,epochs):
        self.w1,self.w2,self.bias = self.binary_crossentropy(x,y,epochs)
        print(f'final weight1 :{self.w1} final weight2:{self.w2} final bias:{self.bias}')
        
    def binary_crossentropy(self,x,y,epochs):
        
        w1=w2=1
        bias=0
        n = len(x)
        learning_rate = 0.5
        
        for i in range(epochs):
            weight_sum = self.w1*(x['age']) + self.w2*(x['affordibility']) + bias
        
            y_predict = self.np_sigmoid(weight_sum)
            loss = log_less(y,y_predict)
            
            w1d = (1/n)*np.dot(np.transpose(x['age']),y-y_predict)
            w2d = (1/n)*np.dot(np.transpose(x['affordibility']),y-y_predict)
            biasd = np.mean(y-y_predict)
            
            self.w1 = self.w1-(learning_rate)*w1d
            self.w2 = self.w2-(learning_rate)*w2d
            self.bias = self.bias-(learning_rate)*biasd
            
            print(f'New weight1 :{self.w1} New weight2:{self.w2} New bias:{self.bias}')
        
        return self.w1, self.w2, self.bias
```
